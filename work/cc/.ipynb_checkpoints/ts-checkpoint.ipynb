##Part 1: Time Series Analysis of SP500 Historical Data

```python
%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt  
import statsmodels.api as sm   
import pandas as pd
import pandas.io.data
import datetime
import seaborn as sb  
sb.set_style('darkgrid')
```

```python
# choose the start and end date
start=datetime.datetime(1993,1,1)
end=datetime.datetime(2015,11,30)
```

```python
# load the sp500 information
sp=pandas.io.data.get_data_yahoo('^GSPC',start,end)
```

```python
# plot difference of Adj Close
sp['Diff'] = sp['Adj Close'] - sp['Adj Close'].shift()  
sp['Diff'].plot(figsize=(12, 8))
```

```python
sp['Log'] = sp['Adj Close'].apply(lambda x: np.log(x))  
sp['LogDiff']=sp['Log']-sp['Log'].shift()
sp['LogDiff'].plot(figsize=(12, 8))  
```

```python
import warnings
warnings.filterwarnings('ignore')
fin_aic= np.inf
for p in range(1,5,1):
    for d in range(2):
        for q in range(1,5,1):
            try:
                arima_model=sm.tsa.ARIMA(sp['LogDiff'].iloc[1:],order=(p,d,q)).fit()
                cur_aic=arima_model.aic
                if cur_aic < fin_aic:
                    fin_aic = cur_aic
                    fin_order = (p, d, q)
            except:
                pass
```

```python
arima_final=sm.tsa.ARIMA(sp['LogDiff'].iloc[1:],order=fin_order).fit()
resid=arima_final.resid
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(resid.values.squeeze(), lags=30, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(resid, lags=30, ax=ax2)
```

```python
sp['ForecastDiff'] = arima_final.fittedvalues  
sp[['LogDiff', 'ForecastDiff']].plot(figsize=(12, 8)) 
```

##Part 2: Technical Analysis on SP500 Historical Data

```python
sp['ud']=sp['Adj Close']>sp['Adj Close'].shift()
features=['Open','High','Low','Volume','Adj Close']
features1=['open1','high1','low1','volume1','aclose1']
features2=['open2','high2','low2','volume2','aclose2']
features3=['open3','high3','low3','volume3','aclose3']
features4=['open4','high4','low4','volume4','aclose4']
```

```python
for i in range(5):
    sp[features1[i]]=sp[features[i]].shift()
    sp[features2[i]]=sp[features[i]].shift(2)
    sp[features3[i]]=sp[features[i]].shift(3)
    sp[features4[i]]=sp[features[i]].shift(4)
df=sp[4:]
```

```python
df['CapCl']=df['Volumel']*df['AClosel']
df['CapOl']=df['Volumel']*df['Openl']
df['CapHl']=df['Volumel']*df['Highl']
df['CapLl']=df['Volumel']*df['Lowl']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import neighbors
from sklearn.svm import SVC

def performClassification(X_train, y_train, X_test, y_test, method, parameters):   
    if method == 'RF':   
        return performRFClass(X_train, y_train, X_test, y_test)
        
    elif method == 'KNN':
        return performKNNClass(X_train, y_train, X_test, y_test)
    
    elif method == 'SVM':   
        return performSVMClass(X_train, y_train, X_test, y_test)
    
def performRFClass(X_train, y_train, X_test, y_test):
    clf = RandomForestClassifier(n_estimators=100, n_jobs=-1)
    clf.fit(X_train, y_train)
    accuracy = clf.score(X_test, y_test)
    return accuracy
        
def performKNNClass(X_train, y_train, X_test, y_test):
    clf = neighbors.KNeighborsClassifier()
    clf.fit(X_train, y_train)
    accuracy = clf.score(X_test, y_test)
    return accuracy

def performSVMClass(X_train, y_train, X_test, y_test):
    clf = SVC()
    clf.fit(X_train, y_train)
    accuracy = clf.score(X_test, y_test)
    return accuracy
```

```python
def divide_df(dataset, start_test):
    X=dataset[['open1','high1','low1','volume1','aclose1',
               'open2','high2','low2','volume2','aclose2']]
    y=dataset.ud
    X_train = X[X.index < start_test]
    y_train = y[y.index < start_test]    
    X_test = X[X.index >= start_test]    
    y_test = y[y.index >= start_test]
    return X_train, y_train, X_test, y_test    
```

```python
X_train, y_train, X_test, y_test=divide_df(df,'2015-1-1')
```

```python
performRFClass(X_train, y_train, X_test, y_test)
```

```python
performKNNClass(X_train, y_train, X_test, y_test)
```

```python
performSVMClass(X_train, y_train, X_test, y_test)
```

```python
def performCV(X_train, y_train, folds, method, parameters):
    print ''
    print 'Parameters --------------------------------> ', parameters
    print 'Size train set: ', X_train.shape
    k = int(np.floor(float(X_train.shape[0])/folds))
    print 'Size of each fold: ', k
    acc = np.zeros(folds-1)
    for i in range(2, folds+1):
        print ''
        split = float(i-1)/i
        print 'Splitting the first ' + str(i) + ' chuncks at ' + str(i-1) + '/' + str(i) 
        data = X_train[:(k*i)]
        output = y_train[:(k*i)]
        print 'Size of train+test: ', data.shape
        index = int(np.floor(data.shape[0]*split))
        X_tr = data[:index]        
        y_tr = output[:index]
        
        X_te = data[(index+1):]
        y_te = output[(index+1):]        
        
        acc[i-2] = performClassification(X_tr, y_tr, X_te, y_te, method, parameters)
        print 'Accuracy on fold ' + str(i) + ': ', acc[i-2]
    return acc.mean()   
```
